```python
def conf_matrix(df):
    confusion_matrix = pd.crosstab(df.goal, df.pred_goal, rownames=['Actual'], colnames=['Predicted'])
    return confusion_matrix
```

```python
def display_matrix(df):
    for i in df.step.unique():
        df_temp = df[df.step == i]
        print(f'step: {i}')
        display(conf_matrix(df_temp))
```
